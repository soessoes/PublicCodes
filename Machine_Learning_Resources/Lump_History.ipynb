```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
JSK = pd.read_csv("JSK_Well_Production_Monthly.csv")
JSK['Date'] = pd.to_datetime(JSK['Date'])
JSK = JSK.drop(['Completion','DaysOn'],axis=1)
JSK.head()
```

```python
K = pd.read_csv("K_Well_Production_Monthly.csv")
K['Date'] = pd.to_datetime(K['Date'])
K = K.drop(['Completion','DaysOn'],axis=1)
K.head()
```

```python
JSK_G = JSK.groupby(['Date'])['OilRate','GasRate','WaterRate'].sum()
JSK_G.index.name = 'date'
JSK_G['Date'] = JSK_G.index
JSK_G.index = range(len(JSK_G.index))
JSK_G['Date'] =pd.to_datetime(JSK_G.Date)
JSK_G_S = JSK_G.sort_values('Date')
JSK_G_S.index = range(len(JSK_G.index))
JSK_G_S[1:200]
JSK_G_S.to_csv('JSK_Sorted_Lumped_Production.csv', sep=',', index=False)
JSK_G_S.plot(x='Date', y='WaterRate')
JSK_G_S.plot(x='Date', y='GasRate')
JSK_G_S.plot(x='Date', y='OilRate')
```

```python
K_G = K.groupby(['Date'])['OilRate','GasRate','WaterRate'].sum()
K_G.index.name = 'date'
K_G['Date'] = K_G.index
K_G.index = range(len(K_G.index))
K_G['Date'] =pd.to_datetime(K_G.Date)
K_G_S = K_G.sort_values('Date')
K_G_S.index = range(len(K_G.index))
K_G_S[1:200]
K_G_S.to_csv('K_Sorted_Lumped_Production.csv', sep=',', index=False)
K_G_S.plot(x='Date', y='WaterRate')
K_G_S.plot(x='Date', y='GasRate')
K_G_S.plot(x='Date', y='OilRate')
```
