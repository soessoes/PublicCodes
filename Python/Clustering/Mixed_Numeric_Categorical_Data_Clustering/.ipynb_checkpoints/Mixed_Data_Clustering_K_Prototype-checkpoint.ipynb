# A prototype for clustering mixed type data set (numeric + categorical) 
## Using K-prototype clustering
__Data__: pressure/temperature sensors with three class sizes: small, medium, large spread in different x,y, coordinates

__Reference__: 

[1]. ZHEXUE HUANG - CLUSTERING LARGE DATA SETS WITH MIXED NUMERIC AND CATEGORICAL VALUES:   [Paper's PDF](https://pdfs.semanticscholar.org/d42b/b5ad2d03be6d8fefa63d25d02c0711d19728.pdf)

```python
import numpy as np
from kmodes.kprototypes import KPrototypes
import pandas as pd
import matplotlib.pyplot as plt
from kmodes.kprototypes import KPrototypes
```

```python
# reading a mixed type dataset (numeric + categorical) for pressure/temperature sensors with
# three classes of sizes small, medium, large

data = pd.read_csv('sensors.csv')
print('data with categorical sizes:\n', data.head(10))

convert_size_to_number = {"size":{"small": 10, "medium": 50, "large": 200}}
data_cat = data.replace(convert_size_to_number, inplace = False)
print('\ndata with categorical sizes converted to numbers to be used for' 
      'plot:\nsmall:10, medium:50, large:100\n', data_cat.head(10))
```
