```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
orig = np.load('./intensity_orig.npy')

#Load the data of the heat map 

Heat_map_x = np.load('./Heat_map_x.npy')
Heat_map_y = np.load('./Heat_map_y.npy')
Heat_map_z = np.load('./Heat_map_z.npy')
title = 'Probability of Correct Class'

# Heat_map_2 = (Heat_map+0.77155155)

#plot the heat map in x direction
threshold_x = 0.023

plt.imshow(Heat_map_x, cmap='bwr',vmax=orig[0][1]+threshold_x,vmin=orig[0][1]-threshold_x)
plt.title(title)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=20)
plt.tick_params(labelsize=14)
# plt.axis('equal')
plt.xlim(-2.5, 17.5)
plt.ylim(25, 0)
plt.savefig('Heat_map_x_Sagital.pdf',bbox_inches = 'tight', pad_inches=0.1)
plt.show()

#plot the heat map in y direction
threshold_y = 0.04
plt.imshow(Heat_map_y, cmap='bwr',vmax=orig[0][1]+threshold_y,vmin=orig[0][1]-threshold_y)
plt.title(title)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=20)
plt.tick_params(labelsize=14)
# plt.axis('equal')
plt.xlim(-2.5, 17.5)
plt.ylim(25, 0)
# plt.axis('equal')
plt.savefig('Heat_map_y_Coronal.pdf',bbox_inches = 'tight', pad_inches=0.1)
plt.show()


#plot the heat map in z direction
threshold_y = 0.03
plt.imshow(Heat_map_z, cmap='bwr',vmax=orig[0][1]+threshold_y,vmin=orig[0][1]-threshold_y)
plt.title(title)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=20)
plt.tick_params(labelsize=14)
# plt.axis('equal')
plt.xlim(22, 2.5)
plt.ylim(25, 0)
plt.savefig('Heat_map_z_axial.pdf',bbox_inches = 'tight', pad_inches=0.1)
plt.show()
```
