```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi, sin, cos
```

```python
x = np.linspace(-pi, pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.close('all')
plt.rc('text', usetex=True)

plt.figure(figsize=(12,8))
ax, fig = plt.gca(), plt.gcf()
plt.plot(x, y1, "-.g", x, y2, '--b', linewidth=3)

plt.xlabel(r'$\theta$', {'color': 'k', 'fontsize': 30})
plt.ylabel(r'\bf{$\sin(\theta)$} ', {'color': 'g', 'fontsize': 30})

plt.xlim(x.min(), x.max()*2)
plt.ylim(-2, 2)

plt.title(r'$\sin(\theta)$ and $\cos(\theta)$ functions', {'color': 'k', 'fontsize': 30})
plt.legend(['sin', 'cos'], shadow=True, loc=(0.02, 0.8), handlelength=1.5, fontsize=20)

plt.xticks((-pi, -pi/2, 0, pi/2, pi), ('$-\pi$','$-\pi/2$','0','$\pi/2$','$+\pi$'), color='k', size=20)
plt.yticks((1, -1, 0), ('ymax', 'ymin', '0'), color='k', size=20)
plt.grid()

## add a text that also supports latex, to a plot
plt.text(5, 0.1, r'$\Delta t = 2\pi/100$', {'color': 'b', 'fontsize': 20, 'ha': 'center', 'va': 'center',})
#           'bbox': dict(boxstyle="round", fc="w", ec="k", pad=0.1)}) #to add box


# Right Y-axis labels
plt.text(1.02, 0.5, r"\bf{$\cos(\theta)$}", {'color': 'b', 'fontsize': 30}, horizontalalignment='left',
         verticalalignment='center', rotation=90, clip_on=False, transform=plt.gca().transAxes)

## add equation(s), to a plot
eq = r"\begin{eqnarray*}" + \
     r"|\nabla\phi| &=& 1,\\" + \
     r"\frac{\partial \phi}{\partial t} + U|\nabla \phi| &=& 0 " + \
     r"\end{eqnarray*}"

plt.text(6.2, 0.9, eq, {'color': 'b', 'fontsize': 20}, va="top", ha="right")


xdata, ydata = 0, 1
bbox = dict(boxstyle="round", fc="0.8")
arrowprops = dict(arrowstyle = "->", connectionstyle = "angle,angleA=0,angleB=90,rad=10")
offset = 60
ax.annotate('data = (%.1f, %.1f)'%(xdata, ydata), (xdata, ydata), xytext=(1*offset, -0.5*offset), 
            textcoords='offset points', bbox=bbox, arrowprops=arrowprops, fontsize = 20)

ax.annotate('local min', xy=(-pi/2, -1), xytext=(-pi/2+1, -0.5),
            arrowprops=dict(facecolor='black', shrink=0.05), bbox=bbox, fontsize = 20)


plt.savefig("figure_saved.pdf", bbox_inches='tight')
plt.show()

```
