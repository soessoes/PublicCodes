### Interactive visualizations

```python
from IPython.html.widgets import *
import numpy as np
import matplotlib.pyplot as plt



def pltFunc(f):
    
    x = np.arange(0,1,0.01)
    y1 = np.sin(2*np.pi*x*f)
    y2 = np.cos(2*np.pi*x*f) + np.sin(2*np.pi*x*f/4)



    # linear
    plt.subplot(221)
    plt.plot(x, y1)
    plt.title('linear')
    plt.grid(True)


    # log
    plt.subplot(222)
    plt.plot(x, y2)
    plt.title('log')
    plt.grid(True)






    
interact(pltFunc, f = (1,10,0.1))
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

plt.close('all')



# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(x, y)
axarr[0].set_title('Sharing X axis')
axarr[1].scatter(x, y)


```
