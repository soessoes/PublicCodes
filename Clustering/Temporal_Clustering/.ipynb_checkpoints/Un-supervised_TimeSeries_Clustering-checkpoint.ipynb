## Here we will do un-supervised k-means clustering for time series using 3 approaches of

1\. __Euclidean distance__ of time series

2\. __DTW__: Dynamic Time Warping of time series

3\. __DBA__: Dynamic time warping Barycenter Averaging of time series 

in tslearn.

```python
# import the required packages
import numpy as np
import matplotlib.pyplot as plt
from tslearn.clustering import TimeSeriesKMeans
from tslearn.datasets import CachedDatasets
from tslearn.preprocessing import TimeSeriesScalerMeanVariance, TimeSeriesResampler
```

#### Creating 4 categories of time series:

```python
## generating time series

pi, sin = np.pi, np.sin
n_sample, f, scalar = 200, 0.5, 4
t = np.linspace(-pi,pi,200) 

y1 = scalar*(4/pi)*sin(2*pi*f*t)
y2 = y1 + scalar*(4/(3*pi))*sin(2*pi*3*f*t)

y = scalar*np.sign(sin(2*pi*f*t)) #an actual square wave

m, noise = 1.5, 0.5
y3 =   m*t + noise*np.random.rand(n_sample)
y4 = 1.2*m*t + noise*np.random.rand(n_sample)
y5 =   -m*t + noise*np.random.rand(n_sample)
y6 =  -1.2*m*t + noise*np.random.rand(n_sample)

y7 = scalar*(4/pi)*sin(2*pi*f*t)
y7[0:100] = 0
X_train = np.array([y1,y2,y3,y4,y5,y6,y7])

## plotting

f = plt.figure(figsize=(18,9))
for i in range(len(X_train)):
    plt.plot(t, X_train[i], linewidth = 4)
    plt.xlim([-4.2,4.2])
plt.legend(['y'+str(i+1) for i in range(len(X_train))], fontsize = 18)
plt.xticks((-pi, -pi/2, 0, pi/2, pi), ('$-\pi$','$-\pi/2$','0','$\pi/2$','$+\pi$'), color='k', size=18)
plt.yticks((-6, -3, 0, 3, 6), ('-6', '-3', '0', '3', '6'), color='k', size=18)
plt.grid()
plt.show()
```

#### Clustering based on Euclidean distance of time series:

```python
# Clustering based on Euclidean distance of time series 
n_clust = 4
seed = 0
km = TimeSeriesKMeans(n_clusters = n_clust, verbose = True, random_state = seed)
y_pred = km.fit_predict(X_train)
print('\nCategorical labels for each of the plots y1-y7:', y_pred)

f = plt.figure(figsize=(18,12))
sz = len(y)
for yi in range(n_clust):
    plt.subplot(2, 2, yi + 1)
    for xx in X_train[y_pred == yi]:
        plt.plot(t, xx.ravel(), "b", alpha = 1)
    plt.plot(t, km.cluster_centers_[yi].ravel(), "r:", linewidth = 4.0)
    plt.title("Euclidean distance based clustering", fontsize = 16)
    plt.grid()
    plt.xticks((-pi, -pi/2, 0, pi/2, pi), ('$-\pi$','$-\pi/2$','0','$\pi/2$','$+\pi$'), color='k', size=18)
    plt.yticks((-6, -3, 0, 3, 6), ('-6', '-3', '0', '3', '6'), color='k', size=18)

plt.show()
```

#### Clustering based on DBA of time series:

```python
# Clustering based on DBA of time series 
n_clust = 4
seed = 0
dba_km = TimeSeriesKMeans(n_clusters = n_clust, n_init = 2, metric = "dtw", verbose=True, 
                          max_iter_barycenter=10, random_state=seed)
y_pred = dba_km.fit_predict(X_train)
print('\nCategorical labels for each of the plots y1-y7:', y_pred)

f = plt.figure(figsize=(18,12))
sz = len(y)
for yi in range(n_clust):
    plt.subplot(2, 2, yi + 1)
    for xx in X_train[y_pred == yi]:
        plt.plot(t, xx.ravel(), "b", alpha = 1)
    plt.plot(t, dba_km.cluster_centers_[yi].ravel(), "r:", linewidth = 4.0)
    plt.title("DBA based clustering", fontsize = 16)
    plt.xticks((-pi, -pi/2, 0, pi/2, pi), ('$-\pi$','$-\pi/2$','0','$\pi/2$','$+\pi$'), color='k', size=18)
    plt.yticks((-6, -3, 0, 3, 6), ('-6', '-3', '0', '3', '6'), color='k', size=18)
    plt.grid()
plt.show()
```

#### Clustering based on DTW of time series:

```python
# Clustering based on DTW of time series 
n_clust = 4
seed = 0
sdtw_km = TimeSeriesKMeans(n_clusters = n_clust, metric = "softdtw", metric_params = 
                           {"gamma_sdtw": .01}, verbose = True, random_state = seed)
y_pred = sdtw_km.fit_predict(X_train)
print('\nCategorical labels for each of the plots y1-y7:', y_pred)

f = plt.figure(figsize=(18,12))
sz = len(y)
for yi in range(n_clust):
    plt.subplot(2, 2, yi + 1)
    for xx in X_train[y_pred == yi]:
        plt.plot(t,xx.ravel(), "b", alpha = 1)
    plt.plot(t,sdtw_km.cluster_centers_[yi].ravel(), "r:", linewidth = 4.0)
    plt.title("DTW based clustering", fontsize = 16)
    plt.grid()
    plt.xticks((-pi, -pi/2, 0, pi/2, pi), ('$-\pi$','$-\pi/2$','0','$\pi/2$','$+\pi$'), color='k', size=18)
    plt.yticks((-6, -3, 0, 3, 6), ('-6', '-3', '0', '3', '6'), color='k', size=18)
plt.show()
```
